<span style='font-family: Arial; font-size: 16px;'>Finding the Tangent Plane </span>

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-8, 8, 50)
y = np.linspace(-8, 9, 50)
X, Y = np.meshgrid(x, y)

Z = X**2 + Y**2

ax.plot_surface(X, Y, Z, cmap='Blues', alpha=1)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.set_xticks(np.arange(-8, 9, 2))
ax.set_yticks(np.arange(-8, 9, 2))
ax.set_zticks(np.arange(0, 140, 20))

# Define function
def f(x, y):
    return x**2 + y**2

# Define partial derivatives
def df_dx(x, y):
    return 2*x

def df_dy(x, y):
    return 2*y

# Tanget plane at (x0, y0)
x0, y0 = 2, 4
z0 = f(x0, y0)

# Tangent plane equation: z = z0 + df_dx(x0, y0) * (x - x0) + df_dy(x0, y0) * (y - y0)
X_tan = np.linspace(x0 - 3, x0 + 3, 10)
Y_tan = np.linspace(y0 - 3, y0 + 3, 10)
X_tan, Y_tan = np.meshgrid(X_tan, Y_tan)
Z_tan = z0 + df_dx(x0, y0) * (X_tan - x0) + df_dy(x0, y0) * (Y_tan - y0)

# Plot tangent plane
ax.plot_surface(X_tan, Y_tan, Z_tan, color='orange', alpha=0.7)

# Plot tangent lines
tangent_x = np.linspace(x0 - 3, x0 + 3, 10)
tangent_z_x = f(tangent_x, y0)  # Tangent line along x
ax.plot(tangent_x, np.full_like(tangent_x, y0), tangent_z_x, color='black', linewidth=2)

tangent_y = np.linspace(y0 - 3, y0 + 3, 10)
tangent_z_y = f(x0, tangent_y)  # Tangent line along y
ax.plot(np.full_like(tangent_y, x0), tangent_y, tangent_z_y, color='black', linewidth=2)

ax.text(x0, y0, z0, r"$x=2$", fontsize=15, color="white")


plt.show()
```

<span style="font-family:Arial; font-size:20px;">**Gradient Descent** </span> <br>

Function: $f(x, y)$ <br>
Goal: find minimum of $f(x, y)$ <br>

**Step1:** <br>
&nbsp;&nbsp;&nbsp;&nbsp; Define a learning rate $\alpha$. <br>
&nbsp;&nbsp;&nbsp;&nbsp; Choose a starting point $(x_0, y_0)$. <br>

**Step2:** <br>
&nbsp;&nbsp;&nbsp;&nbsp; Update: $\begin{bmatrix} x_k\\y_k \end{bmatrix} = \begin{bmatrix} x_{k-1}\\y_{k-1} \end{bmatrix} - \alpha \nabla f(x_{k-1}, y_{k-1}) $ <br>

**Step3:** <br>
&nbsp;&nbsp;&nbsp;&nbsp; Repeat Step2 until you are close enough to the true minimum $(x^*, y^*)$.

<span style="font-family:Arial; font-size:20px;">**Regression with a Perception** </span>

Single Layer Neural Network Perception <br>

**Prediction Function:** <br>
$$ \hat{y} = w_1 x_1 + w_2 x_2 + b $$

**Loss Function:** <br>
$$ L(y, \hat{y}) = \frac{1}{2} (y - \hat{y})^2 $$ 

**Main Goal:** <br>
Find $w_1$, $w_2$, $b$ that give $\hat{y}$ with the least error. <br><br>

**To find optimal values for:** $w_1, w_2, b$, You need gradient descent
$$ w_1 \rightarrow w_1 - \alpha \frac{\partial L}{\partial w_1} $$
$$ w_2 \rightarrow w_2 - \alpha \frac{\partial L}{\partial w_2} $$
$$ b \rightarrow b - \alpha \frac{\partial L}{\partial b} $$

<span style="font-family:Arial; font-size:20px;">**Sigmoid Function** </span>
$$ \sigma(z) = \frac{1}{1 + e^{-z}} $$

<span style="font-family:Arial; font-size:20px;">**Classification with a Perception** </span>

**Prediction Function:**
$$ \hat{y} = \sigma(w_1 x_1 + w_2 x_2 + b) $$ 

**Loss Function:**
$$ L(y, \hat{y}) = -yln(\hat{y}) - (1 - y)ln(1 - \hat{y}) $$

**Main Goal:** <br>
Find $w_1, w_2, b$ that give $\hat{y}$ with the least error. <br><br>

**To find optimal values for:** $w_1, w_2, b$, You need gradient descent
$$ w_1 \rightarrow w_1 - \alpha \frac{\partial L}{\partial w_1} $$
$$ w_2 \rightarrow w_2 - \alpha \frac{\partial L}{\partial w_2} $$
$$ b \rightarrow b - \alpha \frac{\partial L}{\partial b} $$

<span style="font-family:Arial; font-size:20px;">**Newton's Method** <span>

**Update Approximation**

$$ \frac{f(x_k)}{x_k - x_{k+1}} = f'(x_k) $$ <br>
$$ \frac{f(x_k)}{f'(x_k)} = x_k - x_{k+1} $$ <br>
$$ x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)} $$ <br><br>

**NM for Optimization**

$$ x_{k+1} = x_k - \frac{g'(x_k)}{(g'(x_k))'} $$

<span style="font-family:Arial; font-size:20px;">**Curvature** </span>

$ \frac{d^2 x}{dt^2} > 0 $,   Concave up or convex, (local) minimum <br>
$ \frac{d^2 x}{dt^2} < 0 $,   Concave down, (local) maximum <br>
$ \frac{d^2 x}{dt^2} = 0 $,   Need more information, inconclusive

<span style="font-family:Arial; font-size:20px;">**Hessian Matrix** <span>

$$ \nabla f_x = \begin{bmatrix} f_{xx}(x, y) \\ f_{xy}(x, y) \end{bmatrix} $$ <br>
$$ \nabla f_y = \begin{bmatrix} f_{yx}(x, y) \\ f_{yy}(x, y) \end{bmatrix} $$ <br>

$$ H = \begin{bmatrix} f_{xx}(x, y) & f_{xy}(x, y) \\ f_{yx}(x, y) & f_yy(x, y) \end{bmatrix}$$

<span style="font-family:Arial; font-size:20px;">**Newton's Method** <span>
    
1 variable:
$$ x_{k+1} = x_k - \frac{f'(x_k)}{f"(x_k)} $$ <br>
$$ x_{k+1} = x_k - f"(x_k)^{-1} f'(x_k) $$ <br>
    
2 variables: <br>
$$ \begin{bmatrix} x_{k+1}\\y_{k+1} \end{bmatrix} = \begin{bmatrix} x_k\\y_k \end{bmatrix} - 
H^{-1}(x_k, y_k) \nabla f(x_k, y_k)$$
